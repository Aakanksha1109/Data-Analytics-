```python
ndf=df
```

```python
X=ndf.drop(['species'],axis=1)
Y=ndf['species']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size =0.2,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
```

```python
gaussian.fit(xtrain, ytrain)
```

```python
ypred = gaussian.predict(xtest)
print(xtrain)
print("------------\n")
print(xtest)
print("------------\n")
print(ytrain)
print("------------\n")
print(ytest)
print("------------\n")
print(ypred)
```

```python
from sklearn.metrics import precision_score,confusion_matrix,accuracy_score,recall_score,classification_report,ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(ytest,ypred)
cm
```

```python
cm=confusion_matrix(ytest,ypred)
cm
```

```python
acc=accuracy_score(ytest,ypred)
print("Accuracy Score: ",acc)
```

```python
err_rate=1-acc
print("Error Rate: ",err_rate)
```

```python
pr= precision_score(ytest,ypred,average='micro')
print("Precision Score: ",pr)
```

```python
rs = recall_score(ytest,ypred,average='micro')
print("Recall Score: ",rs)
```

```python
cr=classification_report(ytest,ypred)
print("Classification Report: ",cr)
```
