```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Function to remove outliers using z-score method
def remove_outliers(df, threshold=3):
    # Calculate z-scores for each column
    z_scores = (df - df.mean()) / df.std()

    # Remove rows with any z-score greater than the threshold
    df_filtered = df[(z_scores < threshold).all(axis=1)]

    return df_filtered

# Assuming dataset contains the Iris dataset
# You can replace this with your actual dataset
dataset_filtered = dataset.groupby('species').apply(remove_outliers)

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(16,9))

# Boxplot for Sepal Length
axes[0,0].set_title("Distribution of Sepal Length")
sns.boxplot(y="sepal_length", x="species", data=dataset_filtered, orient='v', ax=axes[0, 0])

# Boxplot for Sepal Width
axes[0,1].set_title("Distribution of Sepal Width")
sns.boxplot(y="sepal_width", x="species", data=dataset_filtered, orient='v', ax=axes[0, 1])

# Boxplot for Petal Length
axes[1,0].set_title("Distribution of Petal Length")
sns.boxplot(y="petal_length", x="species", data=dataset_filtered, orient='v', ax=axes[1, 0])

# Boxplot for Petal Width
axes[1,1].set_title("Distribution of Petal Width")
sns.boxplot(y="petal_width", x="species", data=dataset_filtered, orient='v', ax=axes[1, 1])

plt.show()
```
